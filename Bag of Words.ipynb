```python
import nltk
import numpy as np
import pandas as pd
```

```python
imdb=pd.read_csv('https://raw.githubusercontent.com/skathirmani/datasets/master/imdb_sentiment.csv')
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
review=''.join(imdb.review)
wc=WordCloud(background_color='white').generate(review)
plt.imshow(wc)
```

Text Cleaning
1.Conveting character to lower case
2.Remove junk character
3.remove commonly used words.


```python
text=imdb.review.str.lower().str.replace('[^a-z ]','')
ntext=[]
for doc in text:
    ntext=doc.split(' ')
```

```python
stopwords=nltk.corpus.stopwords.words('english')
```

```python
stemmer=nltk.porter.PorterStemmer()
```

```python
#Removing stopwords
word_clean=[]
for word in ntext:
    if word not in stopwords:
        word_clean.append(stemmer.stem(word))
```
